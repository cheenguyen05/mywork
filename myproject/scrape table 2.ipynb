```python
url = 'https://www.idp.com/vietnam/ielts/offsite-test-dates/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
schedule_titles = table.find_all('tr')[0]

schedule_titles
```

```python
table_schedule_titles = [title.text for title in schedule_titles]

print(table_schedule_titles)
```

```python
df = pd.DataFrame(columns = table_schedule_titles)

df
```

```python
column_data = table.find_all('tr')[1:]
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
